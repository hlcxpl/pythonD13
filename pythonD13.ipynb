```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
df= pd.read_csv('Fish.csv');
df
```

1. Crea una nueva columna que incluya el volumen de pez, asumiéndolo como un
cilindro. Para un cilindro con una base circular, el área de la base (A) es igual a πr²,
donde "r" es el radio de la base del cilindro, y la altura (h) es la distancia entre las dos
bases circulares.


```python
# Calcular el radio (ancho del pez dividido por 2)
df['Radius'] = df['Width'] / 2

# Calcular la altura (longitud promedio del pez)
df['Height'] = (df['Length1'] + df['Length2'] + df['Length3']) / 3

# Calcular el volumen usando la fórmula del cilindro: V = π * r² * h
df['Volume'] = np.pi * df['Radius'] ** 2 * df['Height']

# Mostrar las primeras filas del DataFrame con la nueva columna "Volume"
df
```

2. Verifica la correlación posible entre las variables, numérica y gráficamente.
¿Aumenta la correlación al añadir el volumen? Explica.

```python
# Calcular la correlación entre las variables originales
correlation_original = df[['Weight', 'Length1', 'Length2', 'Length3', 'Height', 'Width']].corr()

# Calcular la correlación añadiendo la columna "Volume"
correlation_with_volume = df[['Weight', 'Length1', 'Length2', 'Length3', 'Height', 'Width', 'Volume']].corr()

# Mostrar correlación original
print("Correlación entre las variables originales:")
print(correlation_original)

# Mostrar correlación con la columna "Volume"
print("\nCorrelación añadiendo la columna 'Volume':")
print(correlation_with_volume)

```

```python
# Crear un mapa de calor para la correlación original
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_original, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlación entre las variables originales")
plt.show()

# Crear un mapa de calor para la correlación con la columna "Volume"
plt.figure(figsize=(9, 7))
sns.heatmap(correlation_with_volume, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlación añadiendo la columna 'Volume'")
plt.show()
```

En resumen, agregar la columna 'Volume' ha aumentado la correlación entre el peso de los peces y las otras variables, lo que sugiere que el volumen tiene una relación lineal más fuerte con el peso que las dimensiones individuales. Esto indica que el volumen, calculado como un cilindro utilizando las dimensiones del pez, proporciona información adicional y puede ser una mejor predictor del peso en comparación con las dimensiones individuales

3. Construye un modelo de regresión lineal que relacione el volumen y el peso de los
peces. Realiza una breve evaluación del modelo y grafícalo.


```python
# Definir las variables predictoras (X) y la variable de respuesta (y)
X = df["Volume"]
y = df["Weight"]

# Agregar una constante al predictor (intercepto)
X = sm.add_constant(X)

# Crear el modelo de regresión lineal
model = sm.OLS(y, X).fit()

# Graficar la regresión lineal
plt.scatter(df["Volume"], df["Weight"], label="Datos reales")
plt.plot(df["Volume"], model.predict(X), color='red', label="Regresión Lineal")
plt.title("Regresión lineal entre Volumen y Peso de los Peces")
plt.xlabel("Volumen del pescado")
plt.ylabel("Peso del pescado")
plt.legend()
plt.show()

# Impresión de Resultados
print(model.summary())
```

El coeficiente de determinación es 0.924, lo que significa que aproximadamente el 92.4% de la variabilidad en el peso de los peces puede explicarse por la variable Volume. Esto indica una relación fuerte entre el volumen del pez y su peso.

Coeficiente para Volume: El coeficiente para la variable Volume es 0.6090. Esto implica que, en promedio, por cada unidad de aumento en el volumen del pez, el peso aumenta en 0.6090 unidades.

Significancia Estadística: El valor p asociado con la variable Volume es muy cercano a cero (0.000), lo que indica que la relación entre Volume y Weight es estadísticamente significativa. Además, el modelo en su conjunto es estadísticamente significativo, como indicado por el estadístico F.

Intercepto: El modelo incluye un término constante (intercepto) de 26.8283. Este valor representa el peso estimado del pez cuando el volumen es cero. Sin embargo, en el contexto de este problema, el valor del intercepto puede no tener un significado práctico relevante.

En resumen, el modelo de regresión lineal simple proporciona una buena aproximación para predecir el peso de los peces basándose únicamente en su volumen

4. ¿Qué sucede si aplicas el modelo anterior por separado a cada especie de peces?
Explica.

```python
# Separar los datos por especie
especies = df['Species'].unique()

# Diccionario para almacenar los modelos y errores
modelos = {}
errores = {}

# Aplicar el modelo por separado a cada especie
for especie in especies:
    # Filtrar datos para la especie actual
    datos_especie = df[df['Species'] == especie]
    
    # Definir variables predictoras (X) y variable de respuesta (y)
    X = datos_especie['Volume']
    y = datos_especie['Weight']
    
    # Agregar una constante al predictor (intercepto)
    X = sm.add_constant(X)
    
    # Crear y ajustar el modelo de regresión lineal
    modelo = sm.OLS(y, X).fit()
    
    # Almacenar el modelo en el diccionario
    modelos[especie] = modelo
    
    # Imprimir el resumen del modelo
    print(f"Resultados del modelo para la especie: {especie}")
    print(modelo.summary())
    
    # Calcular el Error Cuadrático Medio (MSE)
    predicciones = modelo.predict(X)
    mse = ((y - predicciones) ** 2).mean()
    errores[especie] = mse
    
    # Graficar el modelo
    plt.figure(figsize=(8, 6))
    plt.scatter(X['Volume'], y, color='blue', label='Datos reales')
    plt.plot(X['Volume'], modelo.predict(X), color='red', linewidth=2, label='Modelo de Regresión Lineal')
    plt.xlabel('Volumen')
    plt.ylabel('Peso')
    plt.title(f'Regresión Lineal para {especie}')
    plt.legend()
    plt.show()

# Imprimir los errores
print("Errores Cuadráticos Medios para cada especie:")
print(errores)
```

Errores Cuadráticos Medios para Cada Especie de Pez:

Bream: 4647.35
Roach: 815.82
Whitefish: 6353.19
Parkki: 32.32
Perch: 3211.30
Pike: 23526.05
Smelt: 1.83
Los errores cuadráticos medios (MSE) indican cuán bien se ajusta el modelo de regresión lineal simple a los datos de cada especie en términos del peso del pez y su volumen. Valores más bajos de MSE indican un mejor ajuste del modelo a los datos. En este caso, "Smelt" tiene el MSE más bajo (1.83), lo que sugiere un buen ajuste del modelo para esta especie en particular. Por otro lado, "Pike" tiene un MSE significativamente más alto (23526.05), lo que indica que el modelo tiene un ajuste menos preciso para esta especie.

5. Construye un modelo de regresión que relacione dos o más variables con el peso de
los peces. Realiza una breve evaluación del modelo.


```python
# Definir las variables predictoras (X) y la variable de respuesta (y)
X = df[['Length1', 'Length2', 'Length3', 'Width', 'Volume']]
y = df['Weight']

# Agregar una constante al predictor (intercepto)
X = sm.add_constant(X)

# Crear el modelo de regresión lineal múltiple
modelo = sm.OLS(y, X).fit()

# Imprimir el resumen del modelo
print(modelo.summary())

# Graficar el modelo
fig = plt.figure(figsize=(12, 8))
sm.graphics.plot_regress_exog(modelo, 'Volume', fig=fig)
plt.show()
```

El coeficiente de determinación es 0.964, lo que significa que aproximadamente el 96.4% de la variabilidad en el peso de los peces puede explicarse por las variables predictoras incluidas en el modelo. Esto indica una relación muy fuerte entre las variables predictoras y el peso de los peces.

Significancia Estadística: La mayoría de las variables predictoras son estadísticamente significativas en el modelo. Sin embargo, Length2 no es significativa, lo que sugiere que no contribuye significativamente a explicar la variabilidad en el peso.

Intercepto: El modelo incluye un término constante (intercepto) de -74.57. Este valor representa el peso estimado del pez cuando todas las variables predictoras son cero. En el contexto de este problema, el valor del intercepto puede no tener un significado práctico relevante debido a las características biológicas de los peces.

Multicolinealidad: El modelo indica que podría haber multicolinealidad, lo que significa que algunas variables predictoras podrían estar altamente correlacionadas entre sí. Esto puede afectar la interpretación de los coeficientes individuales.

En resumen, el modelo de regresión lineal múltiple proporciona una excelente aproximación para predecir el peso de los peces basándose en las dimensiones físicas y el volumen. La relación es positiva para las dimensiones Length3 y Volume, y negativa para Length1 y Width. La variable Length2 no parece ser relevante en este contexto.
